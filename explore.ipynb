```python
from tqdm import tqdm
# let's have a look at two-D space.
N = 35
M = 30
Nz = 33
Ar = np.linspace(0.8,3.0, N)
w = np.linspace (14,17, M)
posar = np.linspace(14.0, 17.4, Nz)
res = []
# FILEPATH: explore.ipynb
for p in tqdm(np.linspace(14.0, 17.4, Nz)):
    params = np.vstack((np.tile(Ar, M), np.repeat(w, N), np.repeat(p, N * M))).T
    LL = likelihood(params)
    LL = LL.reshape((M, N))
    res.append(LL)
```

```python
import matplotlib.animation as animation
Lmax = np.max(res)
fig = plt.figure()
posar = np.linspace(14.0, 17.4, Nz)
def animate(i):
    plt.clf()
    LL = res[i, :, :]
    plt.imshow(np.exp(LL - Lmax), extent=(Ar.min(), Ar.max(), w.min(), w.max()), vmin=0, vmax=1, aspect='auto')
    plt.plot(1.0, 14.4, 'ro')
    plt.xlabel('A')
    plt.ylabel('w')
    plt.title('pos = %.2f' % posar[i])
    plt.colorbar()

ani = animation.FuncAnimation(fig, animate, frames=Nz, interval=200)

# Save the animation as MP4
ani.save('animation.mp4', writer='ffmpeg')
```

```python
res = np.array(res)
posar = np.linspace(14.0, 17.4, Nz)
p = np.exp(res-res.max())
plt.figure()
plt.imshow(p.sum(axis=0), extent = (Ar.min(), Ar.max(), w.min(), w.max()),aspect='auto')
plt.xlabel('A')
plt.ylabel('w')
plt.figure()
plt.imshow(p.sum(axis=1), extent=(Ar.min(), Ar.max(), posar.min(), posar.max()),aspect='auto')
plt.xlabel('A')
plt.ylabel('p')
plt.show()
plt.imshow(p.sum(axis=2),  extent=(w.min(), w.max(), posar.min(), posar.max()), aspect='auto')
plt.xlabel('w')
plt.ylabel('p')

```

```python
plt.figure()
plt.plot(Ar, p.sum(axis=(0)).sum(axis=0))
plt.xlabel('A')
plt.figure()
plt.plot(w, p.sum(axis=(0)).sum(axis=1))
plt.xlabel('w')
plt.figure()
plt.plot(posar, p.sum(axis=(1)).sum(axis=1))
plt.xlabel('p')
```
